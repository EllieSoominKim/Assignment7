## #18
B-트리 높이 h의 범위는 3 ≤ h ≤ 5 (대략 4~5)  
풀이:최소 높이 (꽉 찬 경우): 각 노드가 1000개 키 → 자식 1001개. 높이 h의 트리는 최대 약 1001^(h-1) × 1000개의 키를 담을 수 있음. 1조 = 10¹² ≤ 1001^h ≈ 10^(3h) → h ≥ 4 (h=4일 때 약 10¹²개 수용 가능).
최대 높이 (최소로 채운 경우): 각 노드 최소 ⌈1000/2⌉ = 500개 자식. 높이 h에서 최소 키 수는 약 2 × 500^(h-1) − 1 ≥ 10¹² → 500^(h-1) ≥ 5×10¹¹ → h ≤ 5 (500^4 ≈ 6.25×10¹⁰, 500^5 ≈ 3.1×10¹³)
